```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import pyne.data
from pyne import enrichment as enr

matplotlib.rc('font', family='serif', size=16)

duc = enr.default_uranium_cascade()
minMstar = pyne.data.atomic_mass(duc.j) + 0.1
maxMstar = pyne.data.atomic_mass(duc.k) - 0.1
mstars = np.linspace(minMstar, maxMstar)
```

```python
lofs = []
for mstar in mstars:
    duc.Mstar = mstar
    casc = enr.solve_symbolic(duc)
    lofs.append(casc.l_t_per_feed)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,6))
plt.plot(mstars, lofs, 'r-o', label='$L_t/F$')
plt.xlabel('$M^*$ [amu]')
plt.ylabel('Total per Feed Flow Rate [unitless]')
lgd = plt.legend(loc=0)
```
